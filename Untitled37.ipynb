```python
from datasets import load_dataset
from transformers import AutoTokenizer, TrainingArguments, Trainer, AutoModelForSeq2SeqLM, DataCollatorForSeq2Seq

# Load the astronomy dataset
astro_dataset = load_dataset("allenai/sciq", split="train[:1000]")  # Adjust dataset size as needed

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("google/mt5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("google/mt5-small")

# Preprocessing function for astronomy questions
def preprocess_astro_function(examples, tokenizer):
    inputs = [f"Question: {q} Context: {c}" for q, c in zip(examples['question'], examples['support'])]
    targets = examples['correct_answer']

    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True, padding='max_length')

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Preprocess the dataset
astro_dataset = astro_dataset.map(lambda examples: preprocess_astro_function(examples, tokenizer), batched=True)

# Define data collator
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results_astro_expert",
    per_device_train_batch_size=2,
    num_train_epochs=1,
    save_strategy="no",
    logging_steps=50,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=astro_dataset,
    data_collator=data_collator,
)

# Train the astronomy expert
trainer.train()
```

```python
import os

# Define the directory to save the model
save_directory = "./my_trained_model"  # Change to your desired directory

# Create the directory if it doesn't exist
os.makedirs(save_directory, exist_ok=True)

# Save the model
trainer.save_model(save_directory)
tokenizer.save_pretrained(save_directory)
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer, TrainingArguments, Trainer, AutoModelForSeq2SeqLM, DataCollatorForSeq2Seq
import os

# Load the linguistics dataset
ling_dataset = load_dataset("universal_dependencies", "en_ewt", split="train")  # ~12,543 examples

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("google/mt5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("google/mt5-small")

# Preprocessing function for linguistic structures
def preprocess_ling_function(examples, tokenizer):
    inputs = [f"Sentence: {text}" for text in examples['text']]
    # Create target labels for each sentence individually to keep consistent length
    targets = [
        " ".join([f"{r}-{h}" for r, h in zip(deprel, head)])  # Join dependencies for each sentence
        for deprel, head in zip(examples['deprel'], examples['head'])
    ]

    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")

    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=128, truncation=True, padding='max_length')

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Preprocess the dataset (apply padding before batching)
# The padding argument is removed. Padding and truncation will be handled by the DataCollator.
ling_dataset = ling_dataset.map(lambda examples: preprocess_ling_function(examples, tokenizer), batched=True)
# Define data collator
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model)

# Define training arguments
training_args = TrainingArguments(
    output_dir="./results_ling_expert",
    per_device_train_batch_size=2,
    num_train_epochs=1,
    save_strategy="no",
    logging_steps=50,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=ling_dataset,
    data_collator=data_collator,
)

# Train the linguistics expert
trainer.train()

# Define the directory to save the model
save_directory = "./my_trained_model"  # Change to your desired directory

# Create the directory if it doesn't exist
os.makedirs(save_directory, exist_ok=True)

# Save the model
trainer.save_model(save_directory)
tokenizer.save_pretrained(save_directory)
```

```python
import os
from datasets import load_dataset
from transformers import AutoTokenizer, TrainingArguments, Trainer, AutoModelForSeq2SeqLM, DataCollatorForSeq2Seq
from evaluate import load
import numpy as np

# Load the OpenBookQA dataset
dataset = load_dataset("openbookqa")
train_dataset = dataset["train"].select(range(1000))  # 1,000 samples for training
eval_dataset = dataset["validation"].select(range(200))  # 200 samples for validation

# Debug: Print dataset structure to verify
print("Sample from train dataset:")
print(train_dataset[0])

# Load the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("google/mt5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("google/mt5-small")

# Preprocessing function for OpenBookQA
def preprocess_openbookqa(examples):
    inputs = []
    targets = []

    for q, choices, answer_key in zip(examples["question_stem"], examples["choices"], examples["answerKey"]):
        try:
            # Verify choices is a dictionary with 'text' and 'label' lists
            if not (isinstance(choices, dict) and 'text' in choices and 'label' in choices):
                raise ValueError(f"Unexpected choices format: {choices}")

            choice_texts = choices['text']
            choice_labels = choices['label']

            # Ensure text and label lists have the same length
            if len(choice_texts) != len(choice_labels):
                raise ValueError(f"Mismatched text and label lengths: {choice_texts}, {choice_labels}")

            # Format choices as "A) text, B) text, ..."
            formatted_choices = [f"{label}) {text}" for label, text in zip(choice_labels, choice_texts)]
            choices_str = ", ".join(formatted_choices)
            input_text = f"Question: {q} Choices: {choices_str}"
            inputs.append(input_text)

            # Map answerKey to the corresponding choice text
            target = None
            #The indentation of this for loop was incorrect and has been fixed
            for label, text in zip(choice_labels, choice_texts):
                if label == answer_key:
                    target = text
                    break
            if target is None:
                raise ValueError(f"Answer key {answer_key} not found in labels: {choice_labels}")
            targets.append(target)

        except Exception as e:
            print(f"Error processing example: {e}")
            inputs.append(f"Question: {q} Choices: Unknown")  # Fallback input
            targets.append("Unknown")  # Fallback target

    # Tokenize inputs
    model_inputs = tokenizer(
        inputs,
        max_length=256,
        truncation=True,
        padding=False
    )

    # Tokenize targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(
            targets,
            max_length=32,
            truncation=True,
            padding=False
        )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
# Apply preprocessing
try:
    train_dataset = train_dataset.map(preprocess_openbookqa, batched=True)
    eval_dataset = eval_dataset.map(preprocess_openbookqa, batched=True)
except Exception as e:
    print(f"Preprocessing failed: {e}")
    raise e

# Remove unnecessary columns
columns_to_keep = ["input_ids", "attention_mask", "labels"]
train_dataset = train_dataset.remove_columns([col for col in train_dataset.column_names if col not in columns_to_keep])
eval_dataset = eval_dataset.remove_columns([col for col in eval_dataset.column_names if col not in columns_to_keep])

# Define data collator
data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model, padding=True)

# Define compute_metrics for evaluation
accuracy_metric = load("accuracy")

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    # Decode predictions and labels
    decoded_preds = tokenizer.batch_decode(np.argmax(predictions, axis=-1), skip_special_tokens=True)
    decoded_labels = tokenizer.batch_decode(labels, skip_special_tokens=True)
    # Compute accuracy (exact match)
    accuracy = accuracy_metric.compute(predictions=decoded_preds, references=decoded_labels)
    return {"accuracy": accuracy["accuracy"]}

# Define training arguments
try:
    # Try modern TrainingArguments with evaluation_strategy
    training_args = TrainingArguments(
        output_dir="./results_openbookqa_expert",
        per_device_train_batch_size=8,
        per_device_eval_batch_size=8,
        num_train_epochs=3,
        learning_rate=2e-5,
        evaluation_strategy="epoch",
        save_strategy="epoch",
        save_total_limit=1,
        logging_steps=50,
        load_best_model_at_end=True,
        metric_for_best_model="accuracy",
        report_to="none",
    )
except TypeError:
    # Fallback for older transformers versions
    print("Falling back to older TrainingArguments without evaluation_strategy")
    training_args = TrainingArguments(
        output_dir="./results_openbookqa_expert",
        per_device_train_batch_size=8,
        per_device_eval_batch_size=8,
        num_train_epochs=3,
        learning_rate=2e-5,
        eval_steps=500,  # Evaluate every 500 steps
        save_steps=500,  # Save every 500 steps
        save_total_limit=1,
        logging_steps=50,
        report_to="none",
    )

# Initialize Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
)

# Train the model
try:
    trainer.train()
except Exception as e:
    print(f"Training failed: {e}")
    raise e

# Save the model and tokenizer
save_directory = "./openbookqa_trained_model"
os.makedirs(save_directory, exist_ok=True)
trainer.save_model(save_directory)
```
